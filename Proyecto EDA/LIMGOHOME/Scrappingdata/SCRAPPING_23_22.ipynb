```python
import lxml
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
from fake_useragent import UserAgent                
ua = UserAgent()
headers = {'User-Agent': ua.random}
```

```python
url = "https://www.transfermarkt.es/valencia-cf/kader/verein/1049/plus/0/galerie/0?saison_id=2022"
```

```python
plantilla23_22 = requests.get(url, headers=headers)
```

```python
plantilla23_22 = BeautifulSoup(plantilla23_22.text, "html.parser")
```

```python
temp23={"Nombre":[],
        "Posicion":[],
        "Edad":[],
        "Nacionalidad":[],
        "Valor":[]
        }
```

SACAR EL VALOR DEL JUGADOR EN EL EQUIPO

```python
for valor in plantilla23_22.find_all("td", class_ = "rechts hauptlink"):
    temp23["Valor"].append(valor.text)
```

```python
temp23
```

SACAR NOMBRE JUGADOR

```python
nombres = plantilla23_22.find_all('img', class_='bilderrahmen-fixed lazy lazy')
for nombre in nombres:
    #jugador24 = nombre["title"]
    temp23["Nombre"].append(nombre["title"])
```

SACAR EDAD JUGADOR

Veo que desde la posicion 1 y cada 4 saco el TD donde esta la edad ya que ese campo no tiene ningún atributo o clase que pueda diferenciar del resto.
Saco la longitud total de el td para saber su longitud y recorrer cada campo de la edad.

```python
edad = plantilla23_22.find_all("td",class_ = "zentriert")

longitud=(len(edad)-4) #Saco el total y en el range le pongo 97 porque es el último valor de la cadena que queremos sacar.

valores_edad = [edad[i].text.strip() for i in range(1, longitud, 4)]
temp23["Edad"]=valores_edad
```

SACAR LA POSICION
Como en el anterior no tenemos una etiqueta ni nada que nos indique el campo de la posicion tenemos que sacarlo
a traves de la posicion de la etiqueta.

```python
posiciones = plantilla23_22.find_all("table",class_ = "inline-table")

longitud_pos = (len(posiciones))
for i in range(longitud_pos):
    #print(posiciones[i].find_all("td")[2].text.strip())
    temp23["Posicion"].append(posiciones[i].find_all("td")[2].text.strip())
```

SACAR NACIONALIDAD

saco el td con la clase, veo que los que tienen doble nacionalidad ocupan td diferentes por eso le decimos que coja todos los "img" y la primera posicion de cada td.

```python
nacionalidades = plantilla23_22.find_all("td", class_='zentriert')
for nacion in nacionalidades:
    img = nacion.find_all("img", class_="flaggenrahmen")
    if img:
        #print(img[0].get("title"))
        temp23["Nacionalidad"].append(img[0].get("title"))
```

```python
temp23
```

```python
temporada23_22 =  pd.DataFrame(temp23)
```

```python
temporada23_22.head(10)
```

```python
temporada23_22.Edad = temporada23_22.Edad.astype(int)
```

```python
temporada23_22.info()
```

```python
temporada23_22.Valor = temporada23_22.Valor.str.replace("[^\d,.]", "", regex=True).str.replace(",", ".").str.extract(r'(\d+.\d+|\d+)').astype(float)
```

Pasamos los valores a millones, sabemos que los valores altos que habian no corresponden al valor en millones ya que sabemos que 
la plantilla del valencia no tiene jugadores tan valiosos.

```python
temporada23_22.loc[temporada23_22.Valor >= 100,"Valor"] = temporada23_22.loc[temporada23_22.Valor >= 100,"Valor"]/1000
```

```python
temporada23_22["Valor millones"] = temporada23_22.Valor 
```

```python
temporada23_22 = temporada23_22.drop(columns=["Valor"])
```
