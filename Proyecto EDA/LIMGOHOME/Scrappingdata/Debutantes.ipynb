```python
import lxml
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
from fake_useragent import UserAgent                
ua = UserAgent()
headers = {'User-Agent': ua.random}
```

```python
url = "https://www.transfermarkt.es/valencia-cf/transferstroeme/verein/1049/plus/2?saisonIdVon=2013&saisonIdBis=2022&zuab=zu&verein_id=6182"
```

```python
jugador = requests.get(url, headers=headers)
```

```python
jugador = BeautifulSoup(jugador.text, "html.parser")
```

```python
debutantes = {
    "Nombre": [],
    "Edad debut":[],
    "Nacionalidad": [],
    "Fecha Primer equipo":[]
}
```

```python
nombres_jug = jugador.find_all('img', class_='bilderrahmen-fixed lazy lazy')
for nombre in nombres_jug:
    #jugador24 = nombre["title"]
    debutantes["Nombre"].append(nombre["title"])



    
   
```

```python
edad_actual = jugador.find_all("td",class_ = "zentriert")

longitud=(len(edad_actual)-1) 

valores_edad = [edad_actual[i].text.strip() for i in range(4, longitud, 2)]

debutantes["Edad debut"] = valores_edad
```

```python
nacionalidades = jugador.find_all("td", class_='zentriert')
primera_vez = True
for nacion in nacionalidades:
    img = nacion.find_all("img")
    if img:
        if primera_vez:
            primera_vez = False
            continue
        #print(img[0].get("title"))
        debutantes["Nacionalidad"].append(img[0].get("title"))
```

```python
debutantes
```

```python
primer_equipo = jugador.find_all("td",class_="rechts")

longitud_inco=(len(primer_equipo))

incorporacion = [primer_equipo[i].get_text(strip=True) for i in range(2, longitud_inco, 3)]

debutantes["Fecha Primer equipo"] = incorporacion
```

```python
debutantes
```

```python
debut_jovenes = pd.DataFrame(debutantes)
```

```python
debut_jovenes["Fecha Primer equipo"] = pd.to_datetime(debut_jovenes["Fecha Primer equipo"], format="mixed")
```
